# 2. Password Criteria

```python
import re

def check_password(password):
  # Check length
  length = len(password) > 8
  # Check if password is alphanumeric and has both lower and uppercase
  alphanumeric = bool(re.search('[0-9]', password)) and length
  alphanumeric = bool(re.search('[a-z]', password)) and alphanumeric
  alphanumeric = bool(re.search('[A-Z]', password)) and alphanumeric
  return alphanumeric

### Actually can just check for character in password, then change to true.
```

```python
check_password('abAc1127848913')
```

# 4. Check which class has highest average score and highest maximum score

```python
def mean(lst):
  return float(sum(lst)/len(lst))

def check(A, B):
  # Get mean of each class
  largerMean = 'A' if mean(A) > mean(B) else 'B'
  meansc = max(mean(A), mean(B))
  # Get max of each class
  largerMax = 'A' if max(A) > max(B) else 'B'
  maxsc = max(max(A), max(B))
  return 'Greater Mean: ' + largerMean + ' at ' \
  + str(meansc) + ', Greater Max: ' + largerMax + ' at ' + str(maxsc)
```

```python
scores1 = [10, 11, 15, 20, 55, 76, 90, 84]
scores2 = [4, 9, 12, 98, 35, 42, 4, 5, 10]
check(scores1, scores2)
```
