````python
def checkSame(input, target):
  A = 0
  B = 0
  for i in range(4):
    if input[i] == target[i]:
      A += 1
    elif input[i] in target:
      B += 1
  return A,B
````

````python
checkSame('9875','9876')
````

````python
def output(A, B, attempts):
  if A == 4:
    return "You Win!!\nAttempts: {}".format(attempts)
  else:
    return "{} A (Bulls) {} B (Cows)".format(A,B)
````

````python
import random
target = []
attempts = 0
for i in range(4):
  target.append(str(random.randint(0,9)))
target = ''.join(target)
while True:
  userInput = input("Please input 4 digits: ")
  A, B = checkSame(userInput, target)
  print(output(A,B,attempts))
  attempts += 1
  if A == 4:
    break
````
