## Discussion 1

Based on the `reverser()` function that we already know, develop a function called `reverseAndRepeat(a_str, num)` that reverses the `a_str` string and repeats each character in `a_str` for num times.  

````python
def reverseAndRepeat(a_str, n):
  if len(a_str) == 1:
    return a_str * n
  else:
    new_str = reverseAndRepeat(a_str[1:], n) + a_str[0] * n
    return new_str
````

````python
myStr = "ABC"
str1 = reverseAndRepeat(myStr, 2)
str2 = reverseAndRepeat(myStr, 3)

print(str1)
print(str2)
````

## Discussion 2

Based on the `reverser()` function that we already know, develop a function called `reverseAndOppositeCase(a_str)` that reverses the `a_str` string and changes each character to its opposite case.  

````python
def reverseAndOppositeCase(a_str):
  if len(a_str) == 1:
    if a_str.isupper():
      return a_str.lower()
    else:
      return a_str.upper()
  else:
    x = ''
    if a_str[0].isupper():
      x = a_str[0].lower()
    else:
      x =  a_str[0].upper()
    new_str = reverseAndOppositeCase(a_str[1:]) + x
    return new_str
````

````python
myStr = "AbCdE"

str1 = reverseAndOppositeCase(myStr)
print(str1)
````

## Discussion 3

Based on the `reverser()` function that we already know, develop a function called `symmetricString(a_str)` that  makes a symmetric string, in which `a_str` is its prefix.  

````python
def reverser(a_str):
  if len(a_str) == 1:
    return a_str
  else:
    new_str = reverser(a_str[1:]) + a_str[0]
    return new_str
  
def symmetricString(a_str):
  next = reverser(a_str)
  return a_str + next
````

````python
myStr = "AbCdE"

str1 = symmetricString(myStr)

print(str1)
````
