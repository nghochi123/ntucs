```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Insert your model function (from previous exercises) here
def model(r,N,q0) :
    # Write your code here.
    array = []
    array.append(q0)
    for i in range(N-1):
        prev = array[i]
        nextv = r*(1-prev)*(prev)
        array.append(nextv)
    return np.array(array)
```

## Non periodic behaviour for $r=3.6$.

In the previous exercise we could see (after some initial iterations) periodic behaviour of the sequence $(q_n)$ for $r=3.1$, $r=3.5$ and $r=3.55$ (with a doubling in period length between each listed $r$-value.

On the other hand, for $r=3.6$ we couldn't see any periodicity.

There could be two reasons for this:
1. Maybe, with $r=3.6$, $(q_n)$ does not have any periodicity.
2. Or maybe, the sequence becomes periodic, but we weren't able to spot it.

If you want, you can try to plot a longer part of the sequence and see if you find any periodicity, but it's actually a (difficult to prove) fact that this sequence does not exhibit any periodic behaviour.

Instead of periodicity, let's look at another aspect in which the behaviour for $r=3.6$ turns out to be fundamentally different from the behaviour for $r=3.55$.

## Sensitive dependence on initial value.

Let's look first at the behaviour for $r=3.55$ if we start with two sligthly different initial values $q_0$, let's say $0.6$ vs. $0.65$:

```python
plt.figure(figsize=(12,8), dpi= 100, facecolor='w', edgecolor='k')

plt.plot(model(3.55,100,0.6),'o-',label='$q_0=0.6$')
plt.plot(model(3.55,100,0.65),'o-',label='$q_0=0.65$')
plt.title('Behaviour for $r=3.55$ with two different initial values')
plt.xlabel('$n$')
plt.ylabel('$q_n$')
plt.legend()
plt.show()
```

Now, let's do the same for $r=3.6$:

```python
plt.figure(figsize=(12,8), dpi= 100, facecolor='w', edgecolor='k')

plt.plot(model(3.6,100,0.6),'o-',label='$q_0=0.6$')
plt.plot(model(3.6,100,0.65),'o-',label='$q_0=0.65$')
plt.title('Behaviour for $r=3.6$ with two different initial values')
plt.xlabel('$n$')
plt.ylabel('$q_n$')
plt.legend()
plt.show()
```

Hmmm, let's try again with $r=3.6$ but starting with two initial values very close to each other, let's say $q_0=0.6$ vs. $q_0=0.6001$:

```python
plt.figure(figsize=(12,8), dpi= 100, facecolor='w', edgecolor='k')

plt.plot(model(3.6,100,0.6),'o-',label='$q_0=0.6$')
plt.plot(model(3.6,100,0.6001),'o-',label='$q_0=0.6001$')
plt.title('Behaviour for $r=3.6$ with two very close initial values')
plt.xlabel('$n$')
plt.ylabel('$q_n$')
plt.legend()
plt.show()
```
